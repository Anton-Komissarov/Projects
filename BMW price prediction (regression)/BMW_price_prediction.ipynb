## Preprocessing

Сделаем one hot encoding для категориальных переменных

```python
# Сделали OHE для категориальной переменной
drop_model = pd.get_dummies(bmw['model'], drop_first=True, prefix='model')

# Объединили назад наши таблички
bmw = pd.concat([bmw.drop('model', axis=1), drop_model], axis=1)
bmw
```

```python
# Сделали OHE для категориальной переменной
drop_transmission = pd.get_dummies(bmw['transmission'], drop_first=True, prefix='transmission')

# Объединили назад наши таблички
bmw = pd.concat([bmw.drop('transmission', axis=1), drop_transmission], axis=1)
bmw
```

```python
# Сделали OHE для категориальной переменной
drop_fuelType = pd.get_dummies(bmw['fuelType'], drop_first=True, prefix='fuelType')

# Объединили назад наши таблички
bmw = pd.concat([bmw.drop('fuelType', axis=1), drop_fuelType], axis=1)
bmw
```

```python
from sklearn.model_selection import train_test_split

# Разбиваем на тренировочную и тестовую выборку

X_train, X_test, y_train, y_test = train_test_split(
    bmw.drop(['price'], axis=1),
    bmw['price'],
    test_size=0.30,
    random_state=42
)

X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

Нормируем данные.

```python
from sklearn.preprocessing import StandardScaler

# объявили скалировщик!
scaler = StandardScaler()

# учим скалировщик скалировать все переменны на трэйне
scaler.fit(X_train[['year', 'mileage', 'tax', 'mpg', 'engineSize']])

# Применяем скалировщик к трэйну
df_train_scale = scaler.transform(X_train[['year', 'mileage', 'tax', 'mpg', 'engineSize']])

# Применяем скалирвощик к тесту
df_test_scale = scaler.transform(X_test[['year', 'mileage', 'tax', 'mpg', 'engineSize']])

# Заменяем значения на отскалированные
X_train[['year', 'mileage', 'tax', 'mpg', 'engineSize']] = df_train_scale
X_test[['year', 'mileage', 'tax', 'mpg', 'engineSize']] = df_test_scale
```

## Подбор моделей

Оттолкнемся от примитивной модели и предскажем просто среднее значение.

```python
import numpy as np

y_mean = np.mean(y_train)             # посчитали среднее 
y_pred_naive = np.ones(len(y_test)) * y_mean  # спрогнозировали им цену машины в тестовой выборке
```

```python
from sklearn import metrics  # подгружаем метрики
```

```python
def mean_absolute_percentage_error(y_true, y_pred): 
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100

mean_absolute_percentage_error(y_test, y_pred_naive)
```

```python
def print_metrics(y_test,y_pred):
    print('MAE:', metrics.mean_absolute_error(np.exp(y_test), np.exp(y_pred)))
    print('RMSE:', np.sqrt(metrics.mean_squared_error(np.exp(y_test), np.exp(y_pred))))
    print('R2:',  metrics.r2_score(y_test, y_pred))
    print('MAPE:', mean_absolute_percentage_error(y_test, y_pred))
    pass
```

```python
print_metrics(y_test, y_pred_naive)
```

### Линейная регрессия.

```python
from sklearn.linear_model import LinearRegression

# Объявили модель
model_regression = LinearRegression()

# Обучили модель на тренировочной выборке 
model_regression.fit(X_train, y_train)

# Сделали прогнозы на тестовой выборке 
y_pred_regr = model_regression.predict(X_test)
```

```python
print_metrics(y_test, y_pred_regr)
```

```python
featureImportance = pd.DataFrame({"feature": bmw.drop('price',axis=1).columns, 
                                  "importance": model_regression.coef_})

featureImportance.set_index('feature', inplace=True)
featureImportance.sort_values(["importance"], ascending=False, inplace=True)
featureImportance["importance"].plot(kind='bar', figsize=(10, 6));
```

### Lasso-регрессия

```python
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV

# Решётка для перебора параметра 
param_grid = {'alpha': [0.001, 0.01, 0.05, 0.1, 0.5, 0.8, 1, 5, 10]}

# Объявили модель 
model_lasso = Lasso() 

# Объявили перебор 
grid_cv_lasso = GridSearchCV(model_lasso, param_grid, cv = 5)
grid_cv_lasso.fit(X_train, y_train)
print('Лучшее значение параметра:', grid_cv_lasso.best_params_)

# Сделали прогнозы
y_pred_lasso = grid_cv_lasso.predict(X_test)
```

```python
print_metrics(y_test, y_pred_lasso)
```

### XGBoost

```python
import xgboost as xgb
```

```python
# Объявляем модель

xgb_r = xgb.XGBRegressor(
    learning_rate=0.01,
    max_depth=30,
    n_estimators=1000,
    objective="reg:linear"
)
```

```python
# Фитим модель
xgb_r.fit(X_train, y_train)
```

```python
# Пробуем предсказать

pred = xgb_r.predict(X_test)
```

```python
print_metrics(y_test, pred)
```

Подберем гиперпараметры модели.

```python
# Решётка для перебора параметра 
param_grid = [
    {'n_estimators': [100, 500, 1000],
     'max_depth': [5, 30, 50],
     'learning_rate': [0.01, 0.1, 0.5]}
]

xgb_regressor = xgb.XGBRegressor()
grid_search = GridSearchCV(xgb_regressor, param_grid, scoring='r2', return_train_score=True)
```

```python
grid_search.fit(X_train, y_train)
```

```python
import joblib

# Сохраним модель с лучшими параметрами для дальнейшего использования
xgb_r_GS = grid_search.best_estimator_
joblib.dump(xgb_r_GS, 'XGB_Regressor.pkl')
```

```python
# Подгрузим модель

xgb_r_GS_ = joblib.load('XGB_Regressor.pkl')
```

```python
grid_search.best_estimator_
```

```python
# Пробуем предсказать

pred_best = xgb_r_GS_.predict(X_test)
```

```python
print_metrics(y_test, pred_best)
```
